####  A function call to get all questions in a DataFrame, including the ```load more``` functionality using a for loop

````python
import pandas as pd
import requests
def more_opt():
    all_options = []
    all_questions = []
    for j in range(1978, 2024):
        for i in range (1, 17):
            url = f"https://myschool.ng/classroom/english-language?exam_year={j}&page={i}"
            response = requests.get(url, params= parameter)
            if response.status_code != 200:
                raise Exception(f"Failed to load page ")
            from bs4 import BeautifulSoup
            html =BeautifulSoup(response.text, 'html.parser')
            class_selected = "question-desc mt-0 mb-3"
            quest = html.find_all('div', {'class' : class_selected})
            option_class = "list-unstyled"
            options = html.find_all('ul', {'class': option_class})
    # quest[1]
            new_quest = [questions.text for questions in quest]
            # # if len(new_quest) != len(options):
            #     raise ValueError(f"Mismatch between number of questions ({len(new_quest)}) and options ({len(options)}) on page {i}")
            all_questions.extend(new_quest)  # Add questions
            
            for i in range(len(options)):
                answer_text = [li.get_text(strip=True)[2:] for li in options[i].find_all('li')]
                all_options.append(answer_text)  
        # if len(all_questions) != len(all_options):
        #     raise ValueError("Mismatch between total questions and options.")
        

    df_questions = pd.DataFrame(all_options)
    df_questions.insert(0, 'Question', all_questions)
    return df_questions

# Run the function
df_questions = more_opt()
````
